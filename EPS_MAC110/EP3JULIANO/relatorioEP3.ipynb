<h2> Transformações Sobre listas Uniformes </h2>

<p>Na primeira parte do EP3, implementei as funções de transformação em uma lista com valores uniformemente distribuídos. Os valores foram gerados através de números randômicos entre 10⁻¹⁵ e 1 (O número 0 não poderia ser incluído para não dar erro em algumas funções). </p>
<h5> Transformação Escalar </h5>
<p> Nessa função do EP3, a lista uniforme tinha seus valores transformados a partir de valores reais 'a' e 'b' inseridos pelo usuário. A função fazia a "translação" dos valores iniciais da lista (aproximadamente entre 0 e 1), para valores seguindo a regra:<i><b> a + U[i]*(b-a+1) </b> </i> , onde U[i] é um elemento da lista uniforme.</p>
<p> O resultado foi uma mudança no eixo X, ou seja, o gráfico foi deslocado ("transaladado"), dependendo dos valores de <i><b> 'a' e 'b' </b></i> . O histograma continuou característico de uma distribuição uniforme. Abaixo segue uma imagem do resultado da transformação escalar, para valores de <i><b> a = 20 e b = 1000 </b></i> </p>
<img  src="http://i.imgur.com/OYQZzrq.png">
<h5> Soma de Vetores </h5>
<p> Nessa função do EP3, duas listas com valores uniformemente distribuídos no intervalo <i><b> ]0,1] </b></i> tinham todos os seus valores somados, 1 a 1. A forma do histograma gerado é uma espécie de pirâmide, já que os valores iniciais estavam entre <i><b> ]0,1] </b></i> , uniformemente distribuídos, a maior parte dos valores está perto do valor 1, e vai decaindo simetricamente tanto para 0, quanto para valores maiores que 1.5, que são  valores mais raros de se obter na soma desses números no intervalo em questão. A imagem abaixo mostra o histograma gerado para este tipo de transformação. </p>
<img  src="http://i.imgur.com/k4thKx6.png">
<h5> Raiz de Vetor </h5>
<p> Nessa função do EP3, uma lista era transformada para que posição recebesse a raiz do valor armazenado anteriormente: <i><b> &radic;U[i] </b></i>. Como os valores inicias de U estão uniformemente distribuídos no intervalo <i><b> ]0,1] </b></i>, e como a raiz de um número nesse intervalo sempre dá um número maior que ele mesmo (exceção para o 1), houve a formação de um histograma com forma "escada", já que todos os valores geram um número maior que o original, a frequência é deslocada para números maiores. É fácil de perceber isto no histograma a seguir. </p>
<img  src="http://i.imgur.com/eTzXzlh.png">
<h5> Transformação Box-Muller</h5>
<p> Nessa função do EP3, se aplica a <a href="https://en.wikipedia.org/wiki/Box%E2%80%93Muller_transform">tranformação Box-Muller</a> em duas listas uniformes no intervalo <i><b> ]0,1] </b></i>, gerando duas listas diferentes. A transformação usa uma fórmula que é melhor explicada no link anterior, mas o fato é que ela gera um histograma de distribuição de Gauss (ou distribuição normal), como mostram as imagens dos dois histogramas gerados por essa função.  </p>

<h5> Distribuição Radial </h5>
<p> Nessa função do EP3, é verificada como fica a distribuição dos pontos médios radialmente. Aqui já aparecem diferenças bastante significantes entre os diferentes métodos de gerar cordas. Quanto mais uniforme for essa distribuição, a frequência de cordas que possuem lado maior que o triângulo equilátero inscrito no círculo se aproxima de 50%.  </p>
<h5> Método 1 </h5>
<p>Inicialmente, no Método 1, temos uma distribuição desigual, na qual a maior parte dos pontos médios está na divisão mais próxima à circunferência. No cálculo da probabilidade, isso influencia na quantidade de vezes que uma corda irá ser maior que o lado do triângulo equilátero inscrito nessa circunferência. Na figura do Método 1 abaixo, temos que a soma das categorias <i><b> C1,C2,C3 e C4 </b></i>(que são as categorias cujos pontos médios geram corda maior que o lado do triângulo equiĺátero inscrito na circunferência) é aproximadamente <i><b> 16700</b></i>, ou seja, <i><b> frequência relativa de 33,4%</b></i>, que é a probabilidade esperada no EP2 para este Método. </p>
